```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import zscore
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

df = pd.read_csv(r"/content/Employee.csv")

missing_values = df.isnull().sum()
print(missing_values)

statistics = df.describe()
print(statistics)

# print(df.dtypes)

df['Education']=df['Education'].astype("category")
df['City']= df['City'].astype('category')
df['LeaveOrNot']=df['LeaveOrNot'].astype('bool')
print(df.dtypes)

df["Age_Zscore"] = zscore(df['Age'].dropna())

plt.figure(figsize=(8,5))
sns.boxplot(x=df['Age'])
plt.title("Box plot of Age")
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(x=df['Age_Zscore'])
plt.title("Box plot of zscore Age Normailzation")
plt.show()

df_encoded = pd.get_dummies(df,columns=['Education','City','Gender','EverBenched'])
print(df_encoded)
```
