```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv("/content/student-data.csv")

statistics = df.describe()
print(statistics)

missing_values= df.isnull().sum()
# print(missing_values)


print("The Shape Of The Data Is = ",df.shape)

# print("The Data Types Of the Variables are -\n",df.dtypes)

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

print(df.columns)
df['Age_Normalised'] = scaler.fit_transform(df[['age']])

df['school'] = df['school'].astype('category')
print("The Data Types Of the Variables are -\n",df.dtypes)

plt.figure(figsize=(8,5))
sns.boxplot(df['Age_Normalised'])
plt.title("Age in The normalised Format")
plt.show()

df_encoded = pd.get_dummies(df,columns=['studytime'])
print("Categorical variables for PG column into quantitative variables - \n",df_encoded)
```
