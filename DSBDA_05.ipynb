```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df= pd.read_csv('/content/employee_data_expanded.csv')
df
print("Mean- ",df[['Salary','Age']].mean())

print("Median - ",df[['Salary','Age']].median())

print("Maximum -",df[['Salary','Age']].min())

grouped = df.groupby('Education')['Salary'].agg(['mean','median','max','min'])
print(grouped)

city_count = df['City'].value_counts()
print(city_count)

plt.figure(figsize=(8,5))
sns.histplot(df['Salary'],bins=20,kde=True,color='olive')
plt.title('Salary Distribution')
plt.ylabel('Count')
plt.xlabel('Salary')
plt.show()
```
