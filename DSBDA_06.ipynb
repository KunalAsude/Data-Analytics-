```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/employees.csv')
df.head()



max = df[['SALARY']].max()
print("This is the Mean for Salary -",max)

cov = df[['SALARY']].cov()
print("\nThis is the COV for Salary -",cov)

std = df[['SALARY']].std()
print("\nThis is the Standard Deviation for Salary -",std)
df.head()

grouped= df.groupby('JOB_ID')['SALARY'].agg(['mean','median','min','max'])
print(grouped)

count = df['MANAGER_ID'].value_counts()
print("The count of the Manager Id is - \n",count)

plt.figure(figsize=(8,5))
sns.histplot(df['SALARY'],bins=20,kde=True,color='olive')
plt.title("Visualization of Salary")
plt.xlabel("Salary")
plt.ylabel("Count")
plt.show()
```
