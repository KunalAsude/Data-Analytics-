```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression


df = pd.read_csv('/content/Experience_Salary.csv')
df=df.fillna(df.mean())

X=df[['Experience']]
y=df['Salary']

model=LinearRegression()
model.fit(X,y)

user_exp =15

pred = model.predict([[user_exp]])
print(f"Predicted salary for 15 years experience: {pred[0]}")

plt.scatter(X,y,color='olive',label="Actual Data")
plt.plot(X,model.predict(X),color='red',label="Regression Line")
plt.scatter(user_exp,pred,color="teal",label="Predicted Data")
plt.grid(color='pink')
plt.legend()
```
