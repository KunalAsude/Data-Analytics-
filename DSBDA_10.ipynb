```python
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score

data={'StudyHours':[29, 15, 33, 48, 39],
                    'Pass':[0,0,1,1,1]}
df =pd.DataFrame(data)
df

X=df[['StudyHours']]
y=df['Pass']

model=LogisticRegression()
model.fit(X,y)

predict_25=model.predict([[25]])
print(predict_25)

predict_45=model.predict([[45]])
print(predict_45)

print("Prediction for 25 hours study:", "Pass" if predict_25 == 1 else "Fail")
print("Prediction for 45 hours study:", "Pass" if predict_45 == 1 else "Fail")

y_pred = model.predict(X)

cm = confusion_matrix(y,y_pred)
print(cm)

tn, fp, fn, tp = cm.ravel()
accuracy = accuracy_score(y, y_pred)
error_rate = 1 - accuracy
precision = precision_score(y, y_pred)
recall = recall_score(y, y_pred)

print("\nConfusion Matrix:\n", cm)
print(f"True Positive (TP): {tp}")
print(f"False Positive (FP): {fp}")
print(f"True Negative (TN): {tn}")
print(f"False Negative (FN): {fn}")
print(f"Accuracy: {accuracy:.2f}")
print(f"Error Rate: {error_rate:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
```
